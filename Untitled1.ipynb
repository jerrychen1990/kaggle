```python
def null_analyse(df):
    tmp = df.isnull().sum().rename('null-num')
    tmp = tmp[tmp>0].sort_values(ascending=False)
    rate_df = (tmp/df.shape[0]).rename('null-rate')
    rs_df = pd.concat([tmp, rate_df], axis=1,)
    return rs_df

df = total_raw
null_df = null_analyse(df)
null_df
mode_list = null_df[null_df["null-rate"] < 0.001].index.values
mode_list


# for col in total_raw.columns:
#     df[df[col].isnull()].sum()

```

```python
def cat_fature_unique_num(df):
    tmp = df.dtypes==np.object
    col_idx= tmp[tmp==True].index
    df = df[col_idx]
    return df.nunique().rename('uniq_num')

df = total_raw
cat_fature_unique_num(df)



```

```python
def base_parse(df, drop_list, cat_list, mode_list):
    df = df.drop(drop_list,axis=1)
    
    for c in mode_list:
        df[c].fillna(df[c].mode()[0], inplace=True)
    
    df = pd.get_dummies(df, columns=cat_list, dummy_na=True)

    return df

df = total_raw
df[target_col] = np.log(df[target_col] + 1)
cat_col_series = cat_fature_unique_num(df)
mode_list
drop_list= ["Id"]

parsed_df  = base_parse(df, drop_list, cat_col_series.index.tolist(), mode_list)
parsed_df.info()
null_analyse(parsed_df)

```

```python
def get_k_most_relevant(df, col ,k):
    return df.corr().nlargest(k, col)[col]

df = total_raw
col = 'LotFrontage'
k = 10
get_k_most_relevant(df, col, k)

```

```python
from sklearn.preprocessing import StandardScaler
def feature_predict(df, col, k=10, invalid_feature_list=[]):

    x_feature = get_k_most_relevant(df, col, k).index.tolist()[1:]

    x_feature = [e for e in x_feature if e not in invalid_feature_list + [col]]
#     print(x_feature)

    train_df = df[df[col].notnull()]
    test_df = df[df[col].isnull()]
    if test_df.shape[0] == 0:
        return df
    
    model = LinearRegression()
    train_x = train_df[x_feature]
    
    train_y = train_df[col]
    sc = StandardScaler()
    train_x = sc.fit_transform(train_x)
    model.fit(train_x,train_y)
#     print(train_x.shape)

    test_x = test_df[x_feature]
#     print(test_x.shape)

    test_x = sc.transform(test_x)
    model.predict(test_x)

    df.loc[df[col].isnull(), col] = model.predict(test_x)
    return df


df = parsed_df
col_list  = ['LotFrontage', 'GarageYrBlt', 'MasVnrArea']
invalid_feature_list = col_list + [target_col]
k = 10

for col in col_list:
    df = feature_predict(df, col, k, invalid_feature_list)

null_analyse(df)




```

```python
def split_with_col(df, col, feature_list = None):
    if not feature_list:
        feature_list = [e for e in df.columns if e != col]
    
    train_df = df[df[col].notnull()]
    test_df = df[df[col].isnull()]
    train_x_df = train_df[feature_list]
    train_y_series = train_df[col]
    test_x_df = test_df[feature_list]
    
    return train_x_df, train_y_series, test_x_df


df = parsed_df
col = target_col
train_x_df, train_y_series, test_x_df = split_with_col(df, col)







```
